```python
class BSTNode:
    def __init__(self, key, indices):
        self.key = key
        self.indices = indices
        self.left = None
        self.right = None
        
```

```python
# A class for inserting new nodes and organize them in BST structure.
class BST:
    def __init__(self):
        self.root = None #root of the tree

        # Insert new nodes into the BST.      
    def insert(self, key, index):
        self.root = self._insert_helper(self.root, key, index)
        
    def _insert_helper(self, node, key, index):
        if node is None:
            return BSTNode(key, [index])
        if key < node.key:
            node.left = self._insert_helper(node.left, key, index)
        elif key > node.key:
            node.right = self._insert_helper(node.right, key, index)
        else:
            node.indices.append(index)
        return node
    
    def search(self, key):
        return self._search_helper(self.root, key)
        
    def _search_helper(self, node, key):
        if node is None or node.key == key:
            return [] if node is None else node.indices
        if key < node.key:
            return self._search_helper(node.left, key)
        else:
            return self._search_helper(node.right, key)
        

        
def build_bst(sequence):
    tree = BST()
    for i in range(len(sequence)):
        tree.insert(sequence[i], i)
    return tree

def dna_sequence_search(larger_sequence, target_sequence):
    tree = build_bst(larger_sequence)
    indices = tree.search(target_sequence[0])
    target_length = len(target_sequence)
    result = []
    for index in indices:
        if larger_sequence[index:index+target_length] == target_sequence:
            result.append(index)
    return result
```
